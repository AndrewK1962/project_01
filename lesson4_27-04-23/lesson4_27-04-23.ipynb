```python
# ~ побитовая инверсия
# В битах xчисло 6 можно представить как 00000110. 
# Инвертированное значение  11111001, 11111001 +1 =  11111010
# то есть 2^8 - 6 = 250 (в десятичной системе)
print (~6 == -(6 + 1))
print (~(2+1) == -((2+1) + 1))
```
